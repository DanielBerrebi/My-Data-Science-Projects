```python
def setLamdas(p1,p2,p3):
    #a function for p3 and p2, set the lamdas 1/3 for each if there all was appered in the training set, 
    #or 1/2 for two if there was only two of them appered in the training set (only the rwo that was in the training set will get the values),
    #or 1 if for the  only one of them that appered in the training set 
    #(the function will be called only if there are at least one of them in the training set)
    lamdas=[0,0,0]
    counter=0
    if(p1>0):
        lamdas[0]=1
        counter+=1
    if(p2>0):
        lamdas[1]=1
        counter+=1
    if(p3>0):
        lamdas[2]=1
        counter+=1
    for i in lamdas:
        i=i/counter
    return lamdas
    

def p4Case(sentence,X_train,y_train):
    #checking if the full sentence is in the training set, if so return the label of the sentence by a majority vote or if there is no sentence in the training set return False
    counterN=0
    counterV=0
    for i in range(0,len(X_train)):
        if sentence==X_train[i]:
            if y_train[i]=='N':
                counterN+=1
            else:
                counterV+=1
    if(counterN+counterV)!=0:
        if(counterV/(counterN+counterV))>0.5:
            return 'V'
        else:
            return 'N'
    else:
        return False
        

def p3Case(sentence,X_train,y_train):
    #checking if the only there three words from sentence is in the training set, if so return the label of the sentence by a calculation of sum of each one of three sub sentences probability to V time his lamda,
    #or if there is no sentence in the training set return False
    p1v=p2v=p3v=0
    p1n=p2n=p3n=0
    p1=p2=p3=0
    for i in range(0,len(X_train)):
        if sentence[0:-1]==X_train[i][0:-1]:
        #check if (V,N1,P) is in the training set
            if y_train[i]=='N':
                p1n+=1
            else:
                p1v+=1
        elif sentence[0]+sentence[2]+sentence[3]==X_train[i][0]+X_train[i][2]+X_train[i][3]:
        #else check if (V,P,N2) is in the training set
            if y_train[i]=='N':
                p2n+=1
            else:
                p2v+=1   
        elif sentence[1]+sentence[2]+sentence[3]==X_train[i][1]+X_train[i][2]+X_train[i][3]:
        #else check if (N1,P,N2) is in the training set
            if y_train[i]=='N':
                p3n+=1
            else:
                p3v+=1     
    if(p1v+p2v+p3v+p1n+p2n+p3n)!=0:
    #checking if there is at least one of them in the training set else return False 
        lamdas=setLamdas(p1v+p1n,p2v+p2n,p3v+p3n)
        #calculating the probability of each one of the three sub sentences to V or if there is no sub sentence of one of them in the training set, his probability will be 0
        if(lamdas[0]>0):
            p1=lamdas[0]*p1v/(p1v+p1n)
        if(lamdas[1]>0):
            p2=lamdas[1]*p2v/(p2v+p2n)
        if(lamdas[2]>0):
            p3=lamdas[2]*p3v/(p3v+p3n)
        if(p1+p2+p3)>0.5:

            return 'V'
        else:
            return 'N'
    else:
        return False



def p2Case(sentence,X_train,y_train):
    #checking if the only there two words from sentence is in the training set, if so return the label of the sentence by a calculation of sum of each one of two sub sentences probability to V time his lamda,
    #or if there is no sentence in the training set return False
    p1v=p2v=p3v=0
    p1n=p2n=p3n=0
    p1=p2=p3=0
    for i in range(0,len(X_train)):
        if sentence[0]+sentence[2]==X_train[i][0]+X_train[i][2]:
        #check if (V,P) is in the training set
            if y_train[i]=='N':
                p1n+=1
            else:
                p1v+=1
        elif sentence[2]+sentence[3]==X_train[i][2]+X_train[i][3]:
        #else check if (P,N2) is in the training set
            if y_train[i]=='N':
                p2n+=1
            else:
                p2v+=1
        elif sentence[1:-1]==X_train[i][1:-1]:
        #else check if (N1,P) is in the training set
            if y_train[i]=='N':
                p3n+=1
            else:
                p3v+=1   
    if(p1v+p2v+p3v+p1n+p2n+p3n)!=0:
    #checking if there is at least one of them in the training set else return False
        lamdas=setLamdas(p1v+p1n,p2v+p2n,p3v+p3n)
        #calculating the probability of each one of the three sub sentences to V or if there is no sub sentence of one of them in the training set, his probability will be 0
        if(lamdas[0]>0):
            p1=lamdas[0]*p1v/(p1v+p1n)
        if(lamdas[1]>0):
            p2=lamdas[1]*p2v/(p2v+p2n)
        if(lamdas[2]>0):
            p3=lamdas[2]*p3v/(p3v+p3n)
        if(p1+p2+p3)>0.5:
            return 'V'
        else:
            return 'N'
    else:
        return False



def p1Case(sentence,X_train,y_train):
    #calculating the probability of the sentence to V or N by a majority vote for each time that P and counting the V's and N's is in the training set
    #if the P of the sentence is not in the training set return False
    counterV=0
    counterN=0
    for i in range(0,len(X_train)):
        if sentence[2]==X_train[i][2]:
            #check if (P) is in the training set
            if y_train[i]=='N':
                counterN+=1
            else:
                counterV+=1
    if(counterN+counterV)!=0:
        if(counterV/(counterN+counterV))>0.5:
            return 'V'
        else:
            return 'N'
    else:
        return False


def TestTheModel(X_train, X_test, y_train, y_test):
    #counting the number of the correct predictions of the test set and return the accuracy by the number of the correct predictions divided by the number of rows of the test set
    from tqdm import tqdm
    correct=0
    for i in tqdm(range(0,len(X_test))):
        p4=p4Case(X_test[i],X_train,y_train)
        if p4!=False:
        #checking if the sentence was in the training set
            if p4==y_test[i]:
                correct+=1
        else:
            p3=p3Case(X_test[i],X_train,y_train)    
            if p3!=False:
            #checking if there are triples with prep in the training set
                if p3==y_test[i]:
                    correct+=1 
            else:
                p2=p2Case(X_test[i],X_train,y_train)
                if p2!=False:
                #checking if there are two words with prep in the training set
                    if p2==y_test[i]:
                        correct+=1   
                else:
                    p1=p1Case(X_test[i],X_train,y_train)
                    if p1!=False:
                        if p1==y_test[i]:
                            correct+=1
                    else:
                        #a case when the pronoun is not in the training set 
                        correct+=0.5
    return correct/len(X_test)


#main function that read the data , split it and train and test and print the accuracy of the model
def main():
    #read the txt file and split it into rows
    with open("pp_examples.txt", 'r') as f:
        text = f.read()
        text = text.split('\n')
    text = [x.split(' ') for x in text]
    text
    y=[]
    X=[]
    for i in range(0,len(text)):
        y.append(text[i][-1])
        X.append(text[i][0:-1])
    #split the data into train and test
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
    print("Accuracy: ",TestTheModel(X_train, X_test, y_train, y_test))
#call the main function
main()

            
 
    
```
